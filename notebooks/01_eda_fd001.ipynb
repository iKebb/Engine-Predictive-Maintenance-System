### Experimental Scenario

Data sets consists of multiple multivariate time series. Each data set is further divided into training and test subsets. Each time series is from a different engine – i.e., the data can be considered to be from a fleet of engines of the same type. Each engine starts with different degrees of initial wear and manufacturing variation which is unknown to the user. This wear and variation is considered normal, i.e., it is not considered a fault condition. There are three operational settings that have a substantial effect on engine performance. These settings are also included in the data. The data is contaminated with sensor noise.

The engine is operating normally at the start of each time series, and develops a fault at some point during the series. In the training set, the fault grows in magnitude until system failure. In the test set, the time series ends some time prior to system failure. The objective of the competition is to predict the number of remaining operational cycles before failure in the test set, i.e., the number of operational cycles after the last cycle that the engine will continue to operate. Also provided a vector of true Remaining Useful Life (RUL) values for the test data.

The data are provided as a zip-compressed text file with 26 columns of numbers, separated by spaces. Each row is a snapshot of data taken during a single operational cycle, each column is a different variable. The columns correspond to:

1)	unit number
2)	time, in cycles
3)	operational setting 1
4)	operational setting 2
5)	operational setting 3
6)	sensor measurement  1
7)	sensor measurement  2
8)	(...)


26 ) sensor measurement  26

> Reference: A. Saxena, K. Goebel, D. Simon, and N. Eklund, “Damage Propagation Modeling for Aircraft Engine Run-to-Failure Simulation”, in the Proceedings of the Ist International Conference on Prognostics and Health Management (PHM08), Denver CO, Oct 2008.

````python
def plot_signals_grid_FD001(df):
  sensor_cols = [col for col in df.columns if "sensor_measure" in col]
  n_sensors = len(sensor_cols)
  n_cols = 3
  n_rows = int(np.ceil(n_sensors / n_cols))

  fig, axes = plt.subplots(n_rows, n_cols, figsize=(18, n_rows * 4), constrained_layout=True)
  axes = axes.flatten()

  for i, col in enumerate(sensor_cols):
    ax = axes[i]
    for unit in df["unit"].unique():
      if unit % 10 == 0:
        ax.plot(
          df[df["unit"] == unit]["RUL"].rolling(10).mean(),
          df[df["unit"] == unit][col].rolling(10).mean(),
          alpha=0.7,
          linewidth= 1.5
        )
    ax.set_xlim(250, 0)
    ax.set_xticks(np.arange(0, 300, 25))
    ax.set_xlabel("Remaining Useful Life")
    ax.set_ylabel(col)
    ax.set_title(col)

  for j in range(n_sensors, len(axes)):
    fig.delaxes(axes[j])

  plt.show()
````

````python
def plot_sensors_boxplot(df):
  sensor_cols = [col for col in df.columns if "sensor" in col]
  n_cols = 3
  n_rows = int(np.ceil(len(sensor_cols) / n_cols))

  fig, axes = plt.subplots(n_rows, n_cols, figsize=(18, n_rows * 4), constrained_layout=True)
  axes = axes.flatten()

  for i, col in enumerate(sensor_cols):
    ax = axes[i]
    ax.boxplot(
      df[col],
      vert=True,
      patch_artist=True,
      boxprops=dict(facecolor="lightgray", edgecolor="black"),
      medianprops=dict(color="red", linewidth=2),
      whiskerprops=dict(linewidth=1.5),
      capprops=dict(linewidth=1.5),
      flierprops=dict(marker="o", markersize=4, alpha=0.5)
    )
    ax.set_title(col)
    ax.set_ylabel("Value")
    ax.grid(True, alpha=0.4)

  for j in range(len(sensor_cols), len(axes)):
    fig.delaxes(axes[j])

  plt.show()
````

The 21 sensors come from a simulation of a commercial turbojet engine with ~90,000 lbf of thrust, using NASA's C-MAPSS simulator. It is designed to reflect real engine measurements and their evolution over time, capturing progressive damage —modeled as an exponential loss of **flow** and **efficiency** in the High-Pressure Compressor (HPC), which is the only degraded module in the FD001 dataset.

The sensors are grouped by function:

- **Total Temperatures (T):** measure thermal energy. `T2`, `T24`, `T30`, and `T50` track exhaust gas temperature `(EGT)`. Among them, `T30` and `T50` (corresponding to sensors 3 and 4) provide the clearest signals sensitive to HPC degradation —the engine requires more fuel to maintain thrust, which increases downstream temperatures (SEA LEVEL).

- **Pressures (P):** `P2`, `P15`, `P30`, and `Ps30` are all pressure measurements, but the most relevant are `P30` and `Ps30`. `P30` (sensor 7) is the *TOTAL PRESSURE AT HPC EXIT*, while `Ps30` (sensor 11) is our key pressure measurement, representing the *STATIC PRESSURE* at the same location.

- **Dimensionless Ratios (environment):** `epr` is the engine pressure ratio. It decreases very slightly with damage, as does `BPR` (Bypass Ratio). Of particular interest are `farB` (fuel-air ratio in the combustor) and `phi` (fuel flow / Ps30), which increase because more fuel injection is needed to maintain thrust as `Ps30`, the HPC pressure measurement, drops.

- **Enthalpy:** (air enthalpy extracted for pneumatic systems) and cooling flows such as `W31` (HPT) and `W23` (LPT). These two increase with damage, likely because the control system raises cold air bleed to protect turbine blades against rising temperatures.

- **Command Signals:** these are pilot-demanded speeds through the `TRA`. They are controller inputs, not engine responses; they are not damage outputs and do not change with degradation.

HPC module degradation is modeled as a simultaneous exponential loss of flow ($f$) and efficiency ($e$):

$$f(t) = 1 - d_f - e^{a_f t^{b_f}}$$

$$e(t) = 1 - d_e - e^{a_e t^{b_e}}$$

where $d$ represents the initial deterioration, and $a$, $b$ control the degradation rate. The engine operates by adjusting fuel and speeds to maintain thrust; this is not directly reflected in the sensors as measurable damage, but rather as the system's response to that damage.

- This is important to note because the failure criterion occurs when the Health Index equals 0, defined as the minimum of the operational margins.

- The exhaust gas temperature (EGT) margin is almost equal to the value obtained from the T50 module, which, as a reminder, is the total temperature recorded in the LPT.

We are gonna keep those sensor measurements that are strongly related to the performance loss on the HPC module.

### Highly Useful Sensors for RUL

> *Directly reflect the engine’s response to HPC damage*

| Sensor              | Real Name                                        | Behavior Under Degradation           | Physical Justification                                                                                                                        |
| ------------------- | ------------------------------------------------ | ------------------------------------ | --------------------------------------------------------------------------------------------------------------------------------------------- |
| `sensor_measure_3`  | **T30** — *Total temperature at HPC outlet*      | **↑ Increases**                      | Lower **HPC** efficiency → more heat generation due to friction. *Not an HPT output*. (Fig. 4: EGT ↑ with loss of efficiency/flow)            |
| `sensor_measure_4`  | **T50** — *Total temperature at LPT outlet*      | **↑ Increases significantly**        | Direct proxy of **EGT** (Fig. 4, 8). Failure occurs when the *EGT margin* → 0.                                                                |
| `sensor_measure_7`  | **P30** — *Total pressure at HPC outlet*         | **↓ Decreases**                      | Loss of flow/efficiency in HPC reduces pressure → lower *stall margin* (Fig. 3).                                                              |
| `sensor_measure_11` | **Ps30** — *Static pressure at HPC outlet*       | **↓ Decreases even more than `P30`** | Used directly in the *limit regulator* to prevent low static pressure (p. 2). More sensitive to damage.                                       |
| `sensor_measure_12` | **phi** — *Ratio of fuel flow to Ps30 (pps/psi)* | **↑ Strong increase**                | Dimensionless measure of **engine load**: drop in `Ps30` forces more fuel injection to maintain thrust. Low sensitivity to flight conditions. |
| `sensor_measure_14` | **NRc** — *Corrected core speed (rpm)*           | **↑ Increases**                      | It is `Nc` normalized to ISA (15 °C, 101.325 kPa). Reflects real core effort. More robust than `Nc` (`sensor_measure_9`).                     |
| `sensor_measure_20` | **W31** — *HPT coolant bleed (lbm/s)*            | **↑ Increases**                      | Cold air bleed to cool HPT under rising `T48`. Part of thermal protection strategy (p. 2).                                                    |
| `sensor_measure_21` | **W32** — *LPT coolant bleed (lbm/s)*            | **↑ Increases**                      | Same idea as `W31`, but for LPT (less critical).                                                                                              |
